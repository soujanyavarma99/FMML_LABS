### **NLTK**
NLTK (or Natural Language Tool Kit) is a commonly used library for processing text. We will use this tool in this lab. Lets first install it.


```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
import re
import numpy
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem import SnowballStemmer
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.corpus import wordnet
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from bs4 import BeautifulSoup

def cleanText(text, lemmatize, stemmer):
    """Method for cleaning text from train and test data. Removes numbers, punctuation, and capitalization. Stems or lemmatizes text."""

    if isinstance(text, float):
        text = str(text)
    if isinstance(text, numpy.int64):
        text = str(text)
    try:
        text = text.decode()
    except AttributeError:
        pass

    soup = BeautifulSoup(text, "lxml")
    text = soup.get_text()
    text = re.sub(r"[^A-Za-z]", " ", text)
    text = text.lower()


    if lemmatize:
        wordnet_lemmatizer = WordNetLemmatizer()

        def get_tag(tag):
            if tag.startswith('J'):
                return wordnet.ADJ
            elif tag.startswith('V'):
                return wordnet.VERB
            elif tag.startswith('N'):
                return wordnet.NOUN
            elif tag.startswith('R'):
                return wordnet.ADV
            else:
                return ''

        text_result = []
        tokens = word_tokenize(text)  # Generate list of tokens
        tagged = pos_tag(tokens)
        for t in tagged:
            try:
                text_result.append(wordnet_lemmatizer.lemmatize(t[0], get_tag(t[1][:2])))
            except:
                text_result.append(wordnet_lemmatizer.lemmatize(t[0]))
        return text_result

    if stemmer:
        text_result = []
        tokens = word_tokenize(text)
        snowball_stemmer = SnowballStemmer('english')
        for t in tokens:
            text_result.append(snowball_stemmer.stem(t))
        return text_result
```

## Section 1.2: BAG OF WORDS

A bag-of-words model, or BoW for short, is a way of extracting features from text for use in modeling, such as with machine learning algorithms.

The approach is very simple and flexible, and can be used in many ways for extracting features from documents.

A bag-of-words is a representation of text that describes the occurrence of words within a document.
It is called a “bag” of words, because any information about the order or structure of words in the document is discarded. The model is only concerned with whether known words occur in the document, not where in the document.

```python
# Functions to convert document(s) to a list of words, with the option of removing stopwords. Returns document-term matrix.

def createBagOfWords(train, test, remove_stopwords, lemmatize, stemmer):
    if remove_stopwords:
        vectorizer = CountVectorizer(analyzer='word', input='content', stop_words=stopwords.words('english'))
    else:
        vectorizer = CountVectorizer(analyzer='word', input='content')

    clean_train = []
    for paragraph in train:
        paragraph_result = cleanText(paragraph, lemmatize, stemmer)
        paragraph = " ".join(str(x) for x in paragraph_result)
        clean_train.append(paragraph)

    clean_test = []
    for paragraph in test:
        paragraph_result = cleanText(paragraph, lemmatize, stemmer)
        paragraph = " ".join(str(x) for x in paragraph_result)
        clean_test.append(paragraph)

    bag_of_words_train = vectorizer.fit_transform(clean_train).toarray()
    bag_of_words_test = vectorizer.transform(clean_test).toarray()
    return bag_of_words_train, bag_of_words_test
```

## Section 1.3: TF-IDF
TF-IDF technique is used to find meaning of sentences consisting of words and cancels out the incapabilities of Bag of Words technique which is good for text classification or for helping a machine read words in numbers.

The number of times a term occurs in a document is called its Term frequency (TF).

 Document frequency is the number of documents in which the word is present.  Inverse DF (IDF) is the inverse of the document frequency which measures the informativeness of term *t*.




```python
def createTFIDF(train, test, remove_stopwords, lemmatize, stemmer):
    if remove_stopwords:
        vectorizer = TfidfVectorizer(analyzer='word', input='content', stop_words=stopwords.words('english'))
    else:
        vectorizer =  TfidfVectorizer(analyzer='word', input='content')

    clean_train = []
    for paragraph in train:
        paragraph_result = cleanText(paragraph, lemmatize, stemmer)
        paragraph = " ".join(str(x) for x in paragraph_result)
        clean_train.append(paragraph)

    clean_test = []
    for paragraph in test:
        paragraph_result = cleanText(paragraph, lemmatize, stemmer)
        paragraph = " ".join(str(x) for x in paragraph_result)
        clean_test.append(paragraph)

    tfidf_train = vectorizer.fit_transform(clean_train).toarray()
    tfidf_test = vectorizer.transform(clean_test).toarray()
    return tfidf_train, tfidf_test
```

# **Section 3: KNN MODEL**

Given below are two KNN models; in the first case we are using Bag-of-Words and in the second case we are using TF-IDF.
Note the different metrics and parameters used in each.

```python
from sklearn import metrics, neighbors
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict

## TASK - 1: Tweak the models below and see results with different parameters and distance metrics.

def bow_knn():
    """Method for determining nearest neighbors using bag-of-words and K-Nearest Neighbor algorithm"""

    training_data = pd.read_csv('reviews.csv')
    X_train, X_test, y_train, y_test = train_test_split(training_data["sentence"], training_data["sentiment"], test_size=0.2, random_state=5)
    X_train, X_test = createBagOfWords(X_train, X_test, remove_stopwords=True, lemmatize=True, stemmer=False)
    # print(X_train)
    knn = neighbors.KNeighborsClassifier(n_neighbors=5, weights='uniform', algorithm='auto', leaf_size=30, p=2, metric='euclidean', metric_params=None, n_jobs=1)

    knn.fit(X_train, y_train)
    predicted = knn.predict(X_test)
    acc = metrics.accuracy_score(y_test, predicted)
    print('KNN with BOW accuracy = ' + str(acc * 100) + '%')

    scores = cross_val_score(knn, X_train, y_train, cv=3)
    print("Cross Validation Accuracy: %0.2f" % (scores.mean()))
    print(scores)
    print('\n')
    return predicted, y_test


def tfidf_knn():
    """Method for determining nearest neighbors using tf-idf and K-Nearest Neighbor algorithm"""

    training_data = pd.read_csv('reviews.csv')
    X_train, X_test, y_train, y_test = train_test_split(training_data["sentence"], training_data["sentiment"],
                                                        test_size=0.2, random_state=5)
    X_train, X_test = createTFIDF(X_train, X_test, remove_stopwords=True, lemmatize=True, stemmer=False)
    # print(X_train)
    knn = neighbors.KNeighborsClassifier(n_neighbors=5, weights='distance', algorithm='brute', leaf_size=30, p=2,
                                         metric='cosine', metric_params=None, n_jobs=1)

    knn.fit(X_train, y_train)
    predicted = knn.predict(X_test)
    acc = metrics.accuracy_score(y_test, predicted)
    print('KNN with TFIDF accuracy = ' + str(acc * 100) + '%')

    scores = cross_val_score(knn, X_train, y_train, cv=3)
    print("Cross Validation Accuracy: %0.2f" % (scores.mean()))
    print(scores)
    return predicted, y_test
```

1.Why does the TF-IDF approach generally result in a better accuracy than Bag-of-Words ?
The TF-IDF (Term Frequency-Inverse Document Frequency) approach generally tends to perform better than the Bag-of-Words (BoW) approach in certain natural language processing (NLP) tasks because TF-IDF takes into account not only the occurrence of words but also their importance in the context of the entire document collection. Here are some reasons why TF-IDF often leads to better accuracy compared to Bag-of-Words:

Term Importance:

TF-IDF considers not only the frequency of a term in a document (TF) but also its importance in the entire corpus (IDF). This means that words that are frequent in a specific document but not common across the entire corpus are assigned higher weights. This helps in capturing the discriminative power of words.
Common Word Penalization:

TF-IDF penalizes common words that appear across many documents by giving them lower weights. Words like "the," "is," and "and" are often present in many documents and may not carry much discriminatory information. TF-IDF reduces the influence of such common words.
Dimensionality Reduction:

TF-IDF implicitly performs a form of dimensionality reduction by emphasizing important words and de-emphasizing common ones. This can lead to more compact and informative representations of documents.
Contextual Information:

TF-IDF takes into account the overall context of the document collection. Words that are distinctive to certain documents or topics are assigned higher weights, enabling the model to capture more meaningful patterns.
Noise Reduction:

TF-IDF helps in reducing the impact of noisy or irrelevant words that might be frequent within individual documents but do not contribute much to the overall meaning of the document.
While TF-IDF often outperforms Bag-of-Words in various tasks, it's essential to note that the effectiveness of these methods depends on the specific task and the characteristics of the data. In some cases, especially when dealing with very large datasets or in certain deep learning approaches, more advanced methods like word embeddings (e.g., Word2Vec, GloVe) or transformer-based models (e.g., BERT) may provide even better performance by capturing semantic relationships and contextual information more effectively.

2.Can you think of techniques that are better than both BoW and TF-IDF ?
Certainly! While Bag-of-Words (BoW) and TF-IDF are traditional and effective text representation techniques, there are more advanced methods that have demonstrated superior performance in various natural language processing (NLP) tasks. Here are some techniques that are considered more advanced than BoW and TF-IDF:

Word Embeddings:

Word embeddings, such as Word2Vec, GloVe, and FastText, represent words as continuous vector spaces where semantically similar words are closer together. These embeddings capture semantic relationships and contextual information, providing a more nuanced representation of words and documents.
BERT (Bidirectional Encoder Representations from Transformers):

BERT is a transformer-based model that pre-trains on a large corpus of text, capturing bidirectional context for each word. BERT embeddings have been shown to be highly effective for various NLP tasks, and fine-tuning BERT for specific tasks often leads to state-of-the-art results.
ULMFiT (Universal Language Model Fine-tuning):

ULMFiT is a transfer learning approach for NLP that involves pre-training a language model on a large corpus and then fine-tuning it for a specific task. This method has demonstrated excellent performance even with limited labeled data for the target task.
ELMo (Embeddings from Language Models):

ELMo generates contextualized word embeddings by considering the entire sentence's context. These embeddings capture word meanings that can change based on their context within a sentence.
GPT (Generative Pre-trained Transformer):

GPT models, including GPT-3, are powerful language models that can be fine-tuned for various NLP tasks. They leverage transformer architectures and large-scale pre-training on diverse datasets.
Doc2Vec:

Doc2Vec extends the idea of Word2Vec to entire documents. It represents documents as vectors in a continuous space, allowing for the capture of document-level semantics.
Attention Mechanisms:

Attention mechanisms, as used in transformers, allow models to focus on different parts of the input sequence when making predictions. This enables them to capture long-range dependencies and relationships effectively.
Ensemble Methods:

Ensemble methods combine predictions from multiple models to improve overall performance. Ensemble techniques, such as stacking or bagging, can be applied to various text representation models, including those based on BoW or TF-IDF.
The choice of technique depends on factors such as the task at hand, the amount of labeled data available, and computational resources. State-of-the-art results are often achieved by combining advanced text representation methods with powerful deep learning models in a task-specific manner.

Stemming:
Stemming is a text normalization technique that involves reducing words to their root or base form, known as the stem. The goal is to remove suffixes or prefixes from words, so that similar words map to the same stem. Common stemming algorithms include the Porter stemming algorithm, Lancaster stemming algorithm, and Snowball stemming algorithm.

Pros of Stemming:

Simplicity and Speed: Stemming is computationally less expensive than lemmatization, making it faster and more suitable for real-time applications.

Reduction of Dimensionality: Stemming reduces the number of unique words in a dataset, which can be beneficial for feature space reduction in text-based machine learning models.

Cons of Stemming:

Over-stemming and Loss of Meaning: Stemming can lead to over-stemming, where different words are reduced to the same stem even though they have different meanings. This may result in a loss of meaning.

Not Always Linguistically Correct: Stemming algorithms apply heuristic rules and may not always produce linguistically correct stems. The stems generated may not be actual words.

Lemmatization:
Lemmatization is another text normalization technique that involves reducing words to their base or dictionary form, known as the lemma. Unlike stemming, lemmatization considers the context of the word and aims to transform it into a valid lemma.

Pros of Lemmatization:

Linguistic Accuracy: Lemmatization provides linguistically accurate lemmas, ensuring that the transformed words are valid entries in the language's vocabulary.

Better for Information Retrieval and Text Understanding: Since lemmatization retains the actual words in the language, it is often preferred in tasks where semantic meaning and understanding of the text are crucial, such as information retrieval and question-answering systems.

Cons of Lemmatization:

Computational Complexity: Lemmatization is computationally more expensive than stemming, as it involves dictionary lookups and morphological analysis to find the correct lemma.

May Not Reduce Dimensionality as Much as Stemming: While lemmatization provides more accurate results, it may not reduce the dimensionality of the feature space as much as stemming. This can be a consideration in tasks with limited computational resources.

In summary, the choice between stemming and lemmatization depends on the specific requirements of the task. Stemming is faster and simpler but may sacrifice linguistic accuracy, while lemmatization provides more accurate results at the cost of increased computational complexity. The decision should be made based on the trade-off between speed and accuracy in the context of the application.
